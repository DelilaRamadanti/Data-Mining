```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('data_tugas3.csv',sep=';')
data.tail()
```

```python
data.describe()
```

```python
np.sum(data.isnull())
```

```python
data['X3']=data['X3'].fillna(158.581081)
data['X4']=data['X4'].fillna(23.541096)
np.sum(data.isnull())
```

```python
data.to_csv('tugas3_1.csv')
```

```python
data_new=data
data_new.head()
print(data_new.shape)
```

```python
#missing value outlier
from scipy.stats import stats
z = np.abs(stats.zscore(data._get_numeric_data()))
print(z)
data2 = data[(z<3).all(axis=1)]
print(data2.shape)
```

```python
data2.to_csv('tugas3_2.csv')
```

```python
#missing value dan transformasi
x = data.iloc[0:150,0:6] 
x.tail()
```

```python
from sklearn.preprocessing import normalize
xnorm=normalize(x,norm='l2') 
xnorm=pd.DataFrame(xnorm) 
xnorm.tail()
```

```python
xnorm.to_csv('tugas3_3.csv')
```

```python
print(xnorm.shape)
```
