```python
data['CO_GT ']=data['CO_GT '].fillna(-34.186315)
data['konsentrasi_CO ']=data['konsentrasi_CO '].fillna(316.634587)
data['konsentrasi_NMHC']=data['konsentrasi_NMHC'].fillna(-159.547264)
data['konsentrasi_NO']=data['konsentrasi_NO'].fillna(159.061906)
data['konsentrasi_NMHC ']=data['konsentrasi_NMHC '].fillna(453.079852)
data['konsentrasi_Nox']=data['konsentrasi_Nox'].fillna(544.627504)
data['konsentrasi_NO2']=data['konsentrasi_NO2'].fillna(58.259584)
data['konsentrasi_NO2 ']=data['konsentrasi_NO2 '].fillna(83.804610)
data['konsentrasi_O3']=data['konsentrasi_O3'].fillna(366.323654)
np.sum(data.isnull())
```

```python
from scipy import stats
z= np.abs(stats.zscore(data._get_numeric_data()))
print(z)
data=data[(z<3).all(axis=1)]
print(data.shape)
```
