```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('weatherAUS.csv',sep=';')
data['RainTomorrow'].replace({'No':0,'Yes':1}, inplace= True)
```

```python
data.head()
```

```python
data.describe()
```

```python
np.sum(data.isnull())
```

```python
data['MinTemp']=data.groupby('RainTomorrow')['MinTemp'].transform(lambda x:x.fillna(x.mean()))
data['MaxTemp']=data.groupby('RainTomorrow')['MaxTemp'].transform(lambda x:x.fillna(x.mean()))
data['Rainfall']=data.groupby('RainTomorrow')['Rainfall'].transform(lambda x:x.fillna(x.mean()))
data['Evaporation']=data.groupby('RainTomorrow')['Evaporation'].transform(lambda x:x.fillna(x.mean())
data['Sunshine']=data.groupby('RainTomorrow')['Sunshine'].transform(lambda x:x.fillna(x.mean()))
data['WindGustSpeed']=data.groupby('RainTomorrow')['WindGustSpeed'].transform(lambda x:x.fillna(x.mean()))
data['WindSpeed3pm']=data.groupby('RainTomorrow')['WindSpeed3pm'].transform(lambda x:x.fillna(x.mean()))
data['WindSpeed9am']=data.groupby('RainTomorrow')['WindSpeed9am'].transform(lambda x:x.fillna(x.mean()))
data['Humidity9am']=data.groupby('RainTomorrow')['Humidity9am'].transform(lambda x:x.fillna(x.mean()))
data['Humidity3pm']=data.groupby('RainTomorrow')['Humidity3pm'].transform(lambda x:x.fillna(x.mean()))
data['Pressure9am']=data.groupby('RainTomorrow')['Pressure9am'].transform(lambda x:x.fillna(x.mean()))
data['Pressure3pm']=data.groupby('RainTomorrow')['Pressure3pm'].transform(lambda x:x.fillna(x.mean()))
data['Cloud9am']=data.groupby('RainTomorrow')['Cloud9am'].transform(lambda x:x.fillna(x.mean()))
data['Cloud3pm']=data.groupby('RainTomorrow')['Cloud3pm'].transform(lambda x:x.fillna(x.mean()))
data['Temp9am']=data.groupby('RainTomorrow')['Temp9am'].transform(lambda x:x.fillna(x.mean()))
data['Temp3pm']=data.groupby('RainTomorrow')['Temp3pm'].transform(lambda x:x.fillna(x.mean()))
data['RISK_MM']=data.groupby('RainTomorrow')['RISK_MM'].transform(lambda x:x.fillna(x.mean()))
np.sum(data.isnull())
```

```python
from scipy import stats
z= np.abs(stats.zscore(data._get_numeric_data()))
print(z)
data=data[(z<3).all(axis=1)]
print(data.shape)
```
